# Simple Plotting with Matplotlib

Here we will show some simple plotting using the matplotlib package

```python
import matplotlib.pyplot as plt 
#You can set up packages with abbrievations for ease of use
```

```python
Month = [1,2,3,4,5]
Temperature = [0,6,14,18,25]

plt.figure() #Sets up a figure for you to plot on

plt.plot(Month,Temperature) #plt.plot simple plots the two datasets as a line plot

#Every figure needs axes labels!
plt.xlabel("Month")
plt.ylabel("Temperature")

#Of course you need a title as well
plt.title("Temperature variation with season")
```

Another simple plot is a scatter plot. Let's plot some random numbers! 
You can use numpy to generate arrays of random numbers.


```python
import numpy as np

#np.random.rand generates an array of x numbers between 0 and 1

randoms1 = np.random.rand(50)
randoms2 = np.random.rand(50)

plt.figure()
plt.scatter(randoms1,randoms2)
```

You can also plot multiple things on one plot - this is called a composite plot.

```python
import numpy as np
import matplotlib.pyplot as plt 

#Creating a couple random arrays
randoms1 = np.random.rand(50)*6*np.pi
randoms2 = np.random.rand(50)


x = np.arange(0, 6*np.pi, 0.01) #Generating an array 0-100
cos_function = np.cos(x) #generating a cosine wave, multiplying and shifting it up

plt.figure()
plt.scatter(randoms1,randoms2,color='blue')
plt.plot(x,cos_function,color='red')
```

## 3D plotting

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(8, 8))

#Sets up the figure as a 3D plot
ax = fig.gca(projection='3d')

# Make up some random data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

#Create a Z value from the X and Y values
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm) #Plot the data and add a colormap

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
```
